```python
x1 = xr.DataArray(np.linspace(0,1,11), dims='x', coords={'x':np.linspace(0,1,11)})
x1
```

```python
y1 = np.sin(x1)
y1
```

```python
y1.plot(marker='o')
```

```python
x2 = xr.DataArray(np.linspace(0,1,21), dims='x', coords={'x':np.linspace(0,1,21)})
x2
```

```python
y2 = np.sin(x2)
y2
```

```python
y2.plot(marker='o')
```

```python
ysum = y1 + y2
ysum
```

```python
ysum.plot(marker='o')
```

```python
ysum_int = y1.interp(x=x2) + y2
ysum_int
```

```python
ysum_int.plot(marker='o')
```

```python
x3 = xr.DataArray(np.linspace(0,4*np.pi,201), dims='x', coords={'x':np.linspace(0,4*np.pi,201)})
y3 = np.sin(x3)
```

```python
y3.plot(label='sin')
y3.differentiate('x').plot(label='dsin')
y3.differentiate('x').differentiate('x').plot(label='ddsin')
plt.legend()
```

```python
y3.integrate('x')
```
