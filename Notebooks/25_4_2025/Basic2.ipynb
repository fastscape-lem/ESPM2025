```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
```

```python
x = np.linspace(0,4*np.pi,101)
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x,y,label='sin')
plt.plot(x,np.cos(x), label='cos')
plt.legend()
plt.xlabel('x')
plt.ylabel('periodic function')
plt.title('Test on matplotlib plot')
plt.savefig('test.pdf');
```
